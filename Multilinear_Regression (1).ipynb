```python
import pandas as pd
```

```python
data=pd.read_csv('50_Startups.csv')
data
```

```python
data.info()
```

```python
data.rename(columns={'R&D Spend':'Research_development_Spend','Administration':'Administration','Marketing Spend':'Marketing_Spend','New York':'New_York'},inplace=True)
data
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
data['State']=labelencoder.fit_transform(data['State'])
data
```

```python
data.info()
```

```python
data.corr()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import statsmodels.api as sm
```

```python
y=data['Profit']
y
```

```python
x=data[['Research_development_Spend','Administration','Marketing_Spend','State']]
x
```

```python
# build linear regression model
```

```python
x=sm.add_constant(x)
x
```

```python
model=sm.OLS(y.astype(float),x.astype(float)).fit()
model
```

```python
# calculate cooks distance
```

```python
import numpy as np
np.set_printoptions(suppress=True)
np
```

```python
influence=model.get_influence()
influence
```

```python
#cooks distance
```

```python
cooks=influence.cooks_distance
print(cooks)
```

```python
#visualize cooks distance and influencial points
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Draw plot
plt.figure(figsize = (12, 8))
plt.scatter(data.index, cooks[0])
plt.plot(data.index, cooks[0], color='black')
plt.xlabel('Row Number', fontsize = 12)
plt.ylabel('Cooks Distance', fontsize = 12)
plt.title('Influencial Points', fontsize = 22)
plt.show()
```

```python
# mean cooks distance
```

```python
mean_cooks = np.mean(cooks[0])
mean_cooks
```

```python
mean_cooks_list = [4*mean_cooks for i in data.index]
mean_cooks_list
```

```python
 # Visualize the Influence scores

# Draw plot
plt.figure(figsize = (12, 8))
plt.scatter(data.index, cooks[0])
plt.plot(data.index, mean_cooks_list, color="red")
plt.xlabel('Row Number', fontsize = 12)
plt.ylabel('Cooks Distance', fontsize = 12)
plt.title('Influencial Points', fontsize = 22)
plt.show()
```

```python
influencial_points = data.index[cooks[0] > 4*mean_cooks]
influencial_points
```

```python
# Influencial points
data.iloc[influencial_points, :]
```

```python
# Non Influencial points
noninfluencial_points = data.index[cooks[0] < 4*mean_cooks]
noninfluencial_points
data.iloc[noninfluencial_points, :].head()
```
