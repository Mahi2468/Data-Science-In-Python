```python
import pandas as pd
```

```python
salary=pd.read_csv('Salary_Data.csv')
salary
```

```python
salary.info() # checking for any missing data in dataset
```

```python
# There is no missing data &  There is no character objects
# The dataset is a cleaned dataset,so no need to perform the "EDA"
```

```python
salary.corr()
```

```python
# 97% it is very close to the 1 so, we conclude that we have a strong correlation
# so we conclude that "Simple Linear Regression" works very well
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x='YearsExperience',y='Salary',data=salary,palette='set1')
```

```python
sns.distplot(salary['YearsExperience'])
```

```python
salary.boxplot(column=['YearsExperience'],grid=True)
```

```python
# The data is a left skewed data
```

```python
sns.distplot(salary['Salary'])
```

```python
salary.boxplot(column=['Salary'],grid=True)
```

```python
# The data is a left skewed data
```

```python
# Fitting the linear regression model
```

```python
import statsmodels.formula.api as smf
```

```python
model=smf.ols('Salary~YearsExperience',data=salary).fit()
```

```python
sns.regplot(x='YearsExperience',y='Salary',data=salary)
#To see the best fit line (which is as close as possible to as many points
```

```python
model.params
```

```python
salary.head()
```

```python
model.fittedvalues
```

```python
salary['predicted']=model.fittedvalues
salary[0:5]
```

```python
salary['Error']=model.resid
salary[0:5]
```

```python
salary['Error_sq']=(salary['Error'])**2
salary[0:5]
```

```python
Mean_squared_Error=salary['Error_sq'].mean()
Mean_squared_Error
```

```python
model.rsquared
```

```python
# 95%  data is sufficient to predict the data
# so it is a "Simple Linear Regression"
```
